# Removing the columns with a lot of nulls values, axis =1 its use to show that we actually want to drop the columns

```python
df=crime.drop(labels=['modus_operandi','weapon_code','weapon_description',\
                      'crime_code_2','crime_code_3','crime_code_4','cross_street'],axis=1)
```

```python
# new clean database 
df
```

```python
# sort the data by the date the crime was reported
df.sort_values(by='date_reported')
```

# changing the date columns from object to datetime data type

```python
#Updating datatypes
df['date_reported'] = pd.to_datetime(df['date_reported'], format = '%Y-%m-%d')
df['date_occurred'] = pd.to_datetime(df['date_occurred'])
```

# Finding out the average crime happened on each day 

```python
crime_count_per_day=df.groupby(['date_occurred_no_time', 'crime_description']).size()
```

```python
crime_count_per_day
```

```python
# Finding the average daily crime per crime description
crime_count_per_day.groupby('crime_description').mean() 
```

```python
# resetting the index to turn into a data frame and naming the mean columns
crime_count_per_day.groupby('crime_description').mean().reset_index(name='average_daily_count')
```

```python
# sorting the top 20 crime description sorting by the average 
top_20_crime=crime_count_per_day.groupby('crime_description').mean().reset_index(name='average_daily_count')\
                                                    .sort_values(by='average_daily_count',ascending=False).head(20)
```
